```python
a = 'youtube EZ Python'.isalnum() # 문자열 숫자냐?
b = '123'.isalnum() # 문자열 숫자냐?
c = 'youtube EZ Python'.isalpha() # 문자열 알파벳이냐?? 공백이 있어도 False가 나온다.
d = '123'.isalpha() # 문자열 문자열 알파벳이냐?
e = 'youtube EZ Python'.isascii() # 아스키 코드로 이루어 져있냐?
f = '123'.isascii() # 아스키 코드로 이루어 져있냐?
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
```

```python
a = 'youtube EZ Python'.rjust(30) # 오른쪽
b = 'youtube EZ Python'.ljust(30) # 윈쪽
c ='youtube EZ Python'.center(30) # 가운데
d ='youtube EZ Python'.center(10) # 모자람. 그러면 있는 상태를 손실없이 출력을 함.
print(a) 
print(b)
print(c)
print(d)
```

```python
name = 'William'
age = 10

s = '이름은 %s 입니다. 나이는 %d 입니다.'

print('이름은 %s 입니다. 나이는 %d 입니다.' % (name, age))
print(s % (name, age))
```

```python
print('1. 이름은 {} 입니다. 나이는 {} 입니다.'.format(name, age))
print('2. 이름은 {0} 입니다. 나이는 {0} 입니다.'.format(name, age))
print('3. 이름은 {} 입니다. 나이는 {} 입니다.'.format('youngjin', age))
print('4. 이름은 {name2} 입니다. 나이는 {age2} 입니다.'.format(name2= '!!', age2='11'))
```

```python
print('{} X {} = {}'.format(2,3,6))
print('{0:4} X {1:4} = {2:4}'.format(2,3,6))
print('{0:4} X {1:4} = {2:4}'.format(23456789,3,6))
print('{0:<4} X {1:<4} = {2:<4}'.format(2,3,6))
print('{0:^4} X {1:^4} = {2:^4}'.format(2,3,6))
print('{0:>4} X {1:>4} = {2:>4}'.format(2,3,6))
print('{0:0>4} X {1:0>4} = {2:0>4}'.format(2,3,6))
print('{0:~^4} X {1:!>4} = {2:#>4}'.format(2,3,6))
```

```python
name = 'William'
print(f'1. 제 이름은 {name} 입니다.')
print(f'2. 제 이름은 {name:4} 입니다.')
print(f'3. 제 이름은 {name:<15} 입니다.')
print(f'4. 제 이름은 {name:^15} 입니다.')
print(f'5. 제 이름은 {name:>15} 입니다.')
print(f'6. 제 이름은 {name:!>15} 입니다.')
print(f'7. 제 이름은 {name:~>15} 입니다.')
print(f'8. 제 이름은 {name:0>15} 입니다.')
print(f'9. 제 이름은 {name:#>15} 입니다.')
```

```python
print('{:.3f}'.format(2.1))
print('{0:.4f}'.format(2.1,2.2))
print('{1:.5f}'.format(2.1,2.2))
```

```python
format(999999999999, ',')
print('{0:,.5f}'.format(1234567.123456789))
```

```python
print('{0:010d}'.format(99)) # 10자로 마추겠다.
print('{0:09}'.format(99)) # 10자로 마추겠다.
# print('{0:!9}'.format(99)) # error
print('{0:010.5f}'.format(99)) # 소수점을 포함해서 10자로 마추겠다.
# print('{0:010.5}'.format(99)) error
```

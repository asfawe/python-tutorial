```python
def print_args(*args):
    print(args)
    print(type(args))
    for i in args:
        print(i)
        
print_args(10, True, 'William')
```

```python
def print_args(a,*args):
    print(a)
    print(args)
    print(type(args))
    for i in args:
        print(i)
        
print_args(10, True, 'William')
```

```python
def print_kwargs(**kwargs):
    print(kwargs)
    print(type(kwargs))
    for i in kwargs:
        print(i)

print_kwargs(name='William', age=10)
```
